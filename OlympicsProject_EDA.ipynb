```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
#Data Importing

data = pd.read_csv('athlete_events.csv')
regions = pd.read_csv('noc_regions.csv')
```

```python
#Collecting Information about the two datasets using describe and info function

data.head(5)
```

```python
data.describe()
```

```python
data.info()
```

```python
regions.head()
```

```python
#joining the dataframes into one dataset

merged = pd.merge(data, regions, on='NOC', how='left')
```

```python
merged.head()
```

```python
#distribution of the gold medalists

goldMedals = merged[(merged.Medal == 'Gold')]
goldMedals.head()
```

```python
#checking for null values

goldMedals.isnull().any()
```

```python
#taking only the values that are not null for making plot of the Age to see the distribution

goldMedals = goldMedals[np.isfinite(goldMedals['Age'])]
```

```python
#creating a countplot

plt.figure(figsize=(20, 10))
sns.set()
plt.tight_layout()
sns.countplot(data=goldMedals, x='Age')
plt.title('Distribution of Gold Medals')
```

```python
#Athletes of age greater tha 40

goldMedals['ID'][goldMedals['Age'] > 40].count()
```

```python
 #creating a new dataframe called masterDisciplines in which we will insert the new set of people older than 40 and then creating a visualization with it.

masterDisciplines = goldMedals['Sport'][goldMedals['Age'] > 40]
masterDisciplines.head()
```

```python
plt.figure(figsize=(20, 10))
sns.set()
sns.histplot(data=masterDisciplines)
plt.title('Gold Medals by Discipline')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability if needed
plt.show()
```

```python
#exploring the data of women in athletics

womenInOlympics = merged[(merged.Sex == 'F')]
```

```python
womenInOlympics.head(10)
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(20, 10))
sns.countplot(x='Year', data=womenInOlympics)
plt.title('Women medals per edition of the Games')
```

```python
#Analysis of Medals per Country.
# reviewing the top 5 gold medal countries

goldMedals.region.value_counts().reset_index(name='Medal').head(10)
```

```python
#plotting the above data

totalGoldMedals = goldMedals.region.value_counts().reset_index(name='Medal').head(7)
plt.figure(figsize=(20, 10))
g = sns.catplot(x="index", y="Medal", data=totalGoldMedals,
                height=6, kind="bar", palette="muted")
g.despine(left=True)
g.set_xlabels("Top 5 countries")
g.set_ylabels("Number of Medals")
plt.title('Medals per Country')
```

```python
#Disciplines with the greatest number of Gold Medals

goldMedals.Event.value_counts().reset_index(name='Medal').head(20)
```

```python
#now lets analyse the Variation of male/female athletes over time (Summer Games)

MenOverTime = merged[(merged.Sex == 'M') & (merged.Season == 'Summer')]
WomenOverTime = merged[(merged.Sex == 'F') & (merged.Season == 'Summer')]
```

```python
MenOverTime.head()
```

```python
#for men

part = MenOverTime.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20, 10))
part.loc[:,'M'].plot()
plt.title('Variation of Male Athletes over time')
```

```python
#for women

part = WomenOverTime.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20, 10))
part.loc[:,'F'].plot()
plt.title('Variation of Female Athletes over time')
```

```python
#variation of age along time for male athletes

plt.figure(figsize=(20, 10))
sns.boxplot(x='Year', y='Age', data=MenOverTime)
plt.title('Variation of Age for Male Athletes over time')
```

```python
#variation of age along time for female athletes

plt.figure(figsize=(20, 10))
sns.boxplot(x='Year', y='Age', data=WomenOverTime)
plt.title('Variation of Age for Female Athletes over time')
```

```python
#variation of weight along time for men

plt.figure(figsize=(20, 10))
sns.pointplot(x='Year', y='Weight', data=MenOverTime)
plt.title('Variation of Weight for Male Athletes over time')
```

```python
#variation of weight along time for women

plt.figure(figsize=(20, 10))
sns.pointplot(x='Year', y='Weight', data=WomenOverTime)
plt.title('Variation of Weight for Female Athletes over time')
```

```python
#Variation of height along time for men

plt.figure(figsize=(20, 10))
sns.pointplot(x='Year', y='Height', data=MenOverTime, palette='Set2')
plt.title('Variation of Height for Male Athletes over time')
```

```python
##Variation of height along time for women

plt.figure(figsize=(20, 10))
sns.pointplot(x='Year', y='Height', data=WomenOverTime, palette='Set2')
plt.title('Variation of Height for Female Athletes over time')
```
